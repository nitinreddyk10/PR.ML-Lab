```python
df3=df.where(df['Species']=='Iris-virginica')
df3=df3.dropna()
df3.head()
```

```python
import random
r=df1['Id'].values
r=list(r)
test_1=random.choice(r)
df_t1=df1.where(df1['Id']==test_1)
df1=df1.drop(test_1-1)
print(df1.shape)
df_t1=df_t1.dropna()
df_t1
```

```python
r=df2['Id'].values
r=list(r)
test_2=random.choice(r)
print(test_2)
df_t2=df2.where(df2['Id']==test_2)
df2=df2.drop(test_2-1)
print(df2.shape)
df_t2=df_t2.dropna()
df_t2
```

```python
r=df3['Id'].values
r=list(r)
test_3=random.choice(r)
print(test_3)
df_t3=df3.where(df3['Id']==test_3)
df3=df3.drop(test_3-1)
print(df3.shape)
df_t3=df_t3.dropna()
df_t3
```

```python
m1=df1.mean()
m2=df2.mean()
m3=df3.mean()
print(m1,m2,m3)
```

```python
c1=df1.cov()
c2=df2.cov()
c3=df3.cov()
print(c1,c2,c3)
```

```python
m1=list(m1)
m2=list(m2)
m3=list(m3)
print(m1,m2,m3)
```

```python
r1=df1.iloc[0]
x1=r1.values
x1=list(x1)
x1.remove('Iris-setosa')
print(x1)
```

```python
r2=df2.iloc[0]
x2=r2.values
x2=list(x2)
x2.remove('Iris-versicolor')
print(x2)
```

```python
r3=df3.iloc[0]
x3=r3.values
x3=list(x3)
x3.remove('Iris-virginica')
print(x3)
```

```python
c1=np.array(c1)
c2=np.array(c2)
c3=np.array(c3)
print(c1,c2,c3)
```

```python
from scipy.linalg import inv
def mahalobis_distance(x,c,m):
    sub=np.subtract(x,m)
    subt=sub.T
    C1=inv(c)
    d=np.sqrt(np.dot(np.dot(subt,C1),sub))
    return d
x_t=[x1,x2,x3]
c_t=[c1,c2,c3]
m_t=[m1,m2,m3]
#Calculating for setsosa
d=[]
classes=['Iris-setosa' ,'Iris-versicolor', 'Iris-virginica']
for z in range(len(x_t)):
    d=[]
    for i,j in zip(c_t,m_t):
        dis=mahalobis_distance(x_t[z],i,j)
        d.append(dis)
    d=np.array(d)
    t=z+1
    print(f"Mahalobis distance for {t} sample")
    print(d)
    r=np.argmin(d)
    r=int(r)
    print("predicted class:",classes[r])
    print("original class:",classes[z])
    print("\n")
```
