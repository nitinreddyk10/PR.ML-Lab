```python
import numpy as np 
import pandas as pd
import sympy as sp
```

```python
path='face feature vectors.csv'
rd=pd.read_csv(path)
df=pd.DataFrame(rd)
df
```

```python
x=df.columns
print(x)
```

```python
df=df.drop('Unnamed: 0',axis='columns')
df
```

```python
df_male=df.where(df['Unnamed: 1']=='male')
df_male=df_male.dropna()
df_male
```

```python
df_female=df.where(df['Unnamed: 1']=='female')
df_female=df_female.dropna()
df_female
```

```python
dfmale_train=df_male[:395]
dfmale_test=df_male[395:]
dffemale_train=df_female[:395]
dffemale_test=df_female[395:]
```

```python
df_test=pd.concat([dfmale_test,dffemale_test])
df_test.head()
```

```python
training=[dfmale_train,dffemale_train]
mean_vector=[]
cov_matirx=[]
cov_det=[]
for i in training:
    i=pd.DataFrame(i)
    t=i.drop('Unnamed: 1',axis='columns')
    mean=t.mean()
    mean=list(mean)
    mean_vector.append(mean)
    cov=t.cov()
    cov1=np.array(cov)
    cov2=sp.Matrix(cov1)
    det=cov2.det()
    cov_det.append(det)
    cov_matirx.append(cov1)
```

```python
for i,j in zip(cov_det,mean_vector):
    print(i)
```

```python
def gaussian(x, mean, cov,det):
    d = len(x)
    diff = x - mean
    exponent = -0.5 * np.dot(np.dot(diff.T, np.linalg.inv(cov)), diff)
    return (1.0 / ((2 * np.pi) ** (d / 2) * ((det)**0.5)))* np.exp(exponent)
```

```python
outputs_test=df_test["Unnamed: 1"].values
outputs_test
```

```python
df_test=df_test.drop('Unnamed: 1',axis='columns')
df_test=np.array(df_test)
df_test=list(df_test)
```

```python
def prediction(r):
    x=''
    if(r==0):
        x='male'
    elif(r==1):
        x='female'
    return x

def check(r1,r2):
    acc=0
    for i,j in zip(r1,r2):
        if(i==j):
            acc+=1 
    leng=len(r1)
    return (acc/leng)*100

def check_class(r1,r2):
    acc=[]
    for j in range(2):
        x1=r1[5*j:5*(j+1)]
        x2=r2[5*j:5*(j+1)]
        acc_e=0
        for i,k in zip(x1,x2):
            if(i==k):
                acc_e+=1 
        leng=len(x1)
        t=acc_e/leng 
        t*=100
        acc.append(t)
    return acc[0],acc[1]
def discrminant(a):
    sp.symbols('c')
    c=a
    return sp.log(c)

p=1/2
predictions=[]
discrimininant=[]
for i in range(len(df_test)):
    distri=[]
    for j in range(2):
        t=gaussian(df_test[i],mean_vector[j],cov_matirx[j],cov_det[j])
        t*=p 
        distri.append(t)
    r=np.argmax(distri)
    dis=discrminant(distri[r])
    discrimininant.append(dis)
    pred=prediction(r)
    predictions.append(pred)
acc=check(predictions,outputs_test)
t1,t2=check_class(predictions,outputs_test)
print("Accuracy of the classifer is:",acc)
print("Accuracy of the classifer of male class is:",t1)
print("Accuracy of the classifer of female class is:",t2)
print("Discriminant function:")
for i in discrimininant:
    print(i)
```
