```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path='iris.csv'
rd=pd.read_csv(path)
df=pd.DataFrame(rd)
df
```

```python
outputs=df['Species'].values
outpts=np.array(outputs)
print(np.unique(outputs))
```

```python
df=df.drop('Id',axis='columns')
df.head()
```

```python
#data Normalization
def norm(min,max,t):
  l=((t-min)/(max-min))
  return l 

x=df.columns
x=list(x)
x.pop(-1)
print(x)
for i in x:
  val=df[i].values
  val=np.array(val)
  val_min=np.min(val)
  val_max=np.max(val)
  for j in range(len(val)):
    t=df[i][j]
    l=norm(val_min,val_max,t)
    #to update values in a dataframe
    df.at[j,i]=l
df.head()


```

```python
df1=df.where(df['Species']=='Iris-setosa')
df1=df1.dropna()
df1.head()
```

```python
df2=df.where(df['Species']=='Iris-versicolor')
df2=df2.dropna()
df2.head()
```

```python
df3=df.where(df['Species']=='Iris-virginica')
df3=df3.dropna()
df3.head()
```

```python
df1_train=df1[:40]
df1_test=df1[40:]
df2_train=df2[:40]
df2_test=df2[40:]
df3_train=df3[:40]
df3_test=df3[40:]

```

```python

frames_test=[df1_test,df2_test,df3_test]
```

```python
df_test=pd.concat(frames_test)
df_test.head()
```

```python
training=[df1_train,df2_train,df3_train]
mean_vector=[]
cov_matirx=[]
for i in training:
    i=pd.DataFrame(i)
    t=i.drop('Species',axis='columns')
    mean=t.mean()
    mean=list(mean)
    mean_vector.append(mean)
    cov=t.cov()
    cov=np.array(cov)
    cov_matirx.append(cov)

    

    
```

```python
def gaussian(x, mean, cov):
    d = len(x)
    diff = x - mean
    exponent = -0.5 * np.dot(np.dot(diff.T, np.linalg.inv(cov)), diff)
    return (1.0 / ((2 * np.pi) ** (d / 2) * np.sqrt(np.linalg.det(cov)))) * np.exp(exponent)
```

```python
outputs_test=df_test["Species"].values
outputs_test
```

```python
df_test=df_test.drop('Species',axis='columns')
df_test=np.array(df_test)
df_test=list(df_test)
df_test
```

```python
def prediction(r):
    x=''
    if(r==0):
        x='Iris-setosa'
    elif(r==1):
        x='Iris-versicolor'
    elif(r==2):
        x='Iris-virginica' 
    return x
def check(r1,r2):
    acc=0
    for i,j in zip(r1,r2):
        if(i==j):
            acc+=1 
    leng=len(r1)
    return (acc/leng)*100

def check_class(r1,r2):
    acc=[]
    for j in range(3):
        x1=r1[10*j:10*(j+1)]
        x2=r2[10*j:10*(j+1)]
        acc_e=0
        for i,k in zip(x1,x2):
            if(i==k):
                acc_e+=1 
        leng=len(x1)
        t=acc_e/leng 
        t*=100
        acc.append(t)
    return acc[0],acc[1],acc[2]
def discrminant(a):
    c=a
    return np.log(c)

    

p=1/3
predictions=[]
discrimininant=[]
for i in range(len(df_test)):
    distri=[]
    for j in range(3):
        t=gaussian(df_test[i],mean_vector[j],cov_matirx[j])
        t*=p 
        distri.append(t)
    r=np.argmax(distri)
    dis=discrminant(distri[r])
    discrimininant.append(dis)
    pred=prediction(r)
    predictions.append(pred)
acc=check(predictions,outputs_test)
t1,t2,t3=check_class(predictions,outputs_test)
print("Accuracy of the classifer is:",acc)
print("Accuracy of the classifer of setsosa class is:",t1)
print("Accuracy of the classifer of versicolor class is:",t2)
print("Accuracy of the classifer of virginica class is:",t3)
print("Discriminant function:")
for i in discrimininant:
    print(i)



        
```
